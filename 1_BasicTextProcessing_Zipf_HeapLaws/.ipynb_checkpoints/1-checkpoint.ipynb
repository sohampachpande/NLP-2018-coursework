# NLP Assignment 1: Basic Text Processing and Laws

```python
import nltk
import string
import matplotlib.pyplot as plt
import numpy as np
from nltk.corpus import wordnet as wn

f = open('TomSawyer.txt',encoding="utf8")
corpus = f.read()

corpus = corpus.lower()    # to ensure words with different case for a letter aren't counted twice
corpus = corpus.replace('--', '')
sent_corpus = nltk.sent_tokenize(corpus)    # split sentences
```

### Get Number of Tokens

```python
punctuation_char = string.punctuation + '“”'

corpus = corpus.translate(str.maketrans('','',punctuation_char))
words = nltk.word_tokenize(corpus)
no_of_token = len(words)
print("No. of Tokens= {}".format(no_of_token))
```

### Get Number of Types

```python
word_count = nltk.FreqDist(words)
freq_type = word_count.most_common()
no_of_type = len(freq_type)
print("No. of Types= {}".format(no_of_type))
```

### Type to Token Ratio (TTR)

```python
print("TTR= {}".format(round(no_of_type/no_of_token,5)))
```

### Zipf's Law - Word Position in descending order of frequency vs frequency - f.r = k

```python
list_rank_word, list_freq_word = [], []
for wr in freq_type:
    if wr[1] not in list_freq_word:    # consider only 1 rank for a particular frequency
        list_rank_word.append(freq_type.index(wr))
        list_freq_word.append(wr[1])

plt.xlabel('Frequency of Words')
plt.ylabel('Rank of Words in decreasing order of frequency')
plt.plot(list_freq_word, list_rank_word, 'r')
# plt.annotate("Emperical data",xy=[3000,200],color='red')

plt.show()
```

### Zipf's Law - Length of Words vs frequency - f.l = k'

```python
list_len_word, list_freq_word = [], []
for wlf in freq_type:
    list_len_word.append(len(wlf[0]))
    list_freq_word.append(wlf[1])

plt.xlabel('Frequency of Words')
plt.ylabel('Length of Words')
plt.plot(list_freq_word, list_len_word,'r')
# plt.annotate("Emperical data",xy=[6000,200],color='red')

plt.show()
```

### Zipf's Law - Meanings of Word vs Frequency - m proportional to root of f

```python
list_no_of_meaning, list_freq = [], []
for r in range(100,500,25):
    list_freq.append(freq_type[r][1]**0.5)
    
    count_meanings = 0
    for syn in wn.synsets(freq_type[r][0]):
        for l in syn.lemmas():
            count_meanings += 1
    
    list_no_of_meaning.append(count_meanings)

# print(list_freq,'\n\n', list_no_of_meaning)

# Plot
plt.xlabel('Frequency of Words')
plt.ylabel('No of meanings of word')
plt.plot(list_freq, list_no_of_meaning)

plt.show()
```

### Heaps' Law

```python
# Heaps Law
list_size_vocab, list_no_of_token, word_upto_sent_s = [], [], []
word_upto_sent_s = []
for s in sent_corpus:
    s = s.translate(str.maketrans('','',punctuation_char))
    word_s = nltk.word_tokenize(s)
    
    for w in word_s:
        word_upto_sent_s.append(w)

    # for list_size of vocabulary
    list_no_of_token.append(len(word_upto_sent_s))
    word_count_fd = nltk.FreqDist(word_upto_sent_s)
    no_of_word = len(word_count_fd.most_common())
    list_size_vocab.append(no_of_word)
```

### Obtain parameters using Curve fitting

```python
from scipy.optimize import curve_fit
# function format
def func(n,k,b):
    return k*(n**b)

params_k_b_opt, pcovariance = curve_fit(func, list_no_of_token, list_size_vocab)
print('k= {}, beta= {}'.format(params_k_b_opt[0], params_k_b_opt[1]))
```

### Plot Data for Heaps' Law

```python
# Plot
plt.ylabel('Size of Vocabulary')
plt.xlabel('Number of Tokens')

list_predicted_size_vocab = []
for i in list_no_of_token:
    list_predicted_size_vocab.append(func(i,params_k_b_opt[0],params_k_b_opt[1]))

plt.plot(list_no_of_token, list_predicted_size_vocab, 'b--', label = 'using paramaters from curve fitting') 
plt.plot(list_no_of_token, list_size_vocab, 'r', label = 'Emperical data')

plt.legend()
plt.show()
```
