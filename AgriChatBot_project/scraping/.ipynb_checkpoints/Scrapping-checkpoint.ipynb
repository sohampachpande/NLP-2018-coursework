```python
import requests
import pandas as pd
import string
```

```python
states = {11:"Maharashtra", 18_32:'Tamil Nadu', 15_31: Odisha, 17_35:'Rajasthan',4_28:'Gujarat', 10_45:'MP'}
```

```python
states_c = [10]
district_limits=[45]
statename_dict = {15:'Odisha', 17: 'Rajasthan', 4:'Gujarat', 10:'MadhyaPradesh'}
```

```python
url_datakcc = 'http://dackkms.gov.in/Account/API/kKMS_QueryData.aspx'
for stateCode,districCodeLimit in zip(states_c,district_limits):
    # stateCode = 15
    # districCodeLimit = 32
    for districtCode in range(1,districCodeLimit+1):
        df = pd.DataFrame();
        for yyyy in range(2017,2018):
            for mm in range(1,13):
                params = {'Month':mm, 'Year':yyyy, 'StateCD':str(stateCode).rjust(2,'0'), 'DistrictCd':stateCode*100+districtCode}
                data_response = requests.request('GET',url_datakcc,params=params)
                print(data_response.url)
                data_json=data_response.json()
                try:
                    for d in data_json["data"]:
                        df = df.append(d, ignore_index=True)
                except KeyError:
                    pass
        df.to_csv('./data/{}_{}.csv'.format(stateCode,districtCode))
        df.to_csv('./data_name/{}_{}.csv'.format(statename_dict[stateCode],districtCode))
```

```python
useless_words = ['Ask', 'about', 'information', 'problem', 'for', 'ask','in', 'and','all',
                'ASK','ABOUT','INFORMATION','informations','FOR','THE','asked', 'Asked', 'ASKED']
```

```python
query_text = df['QueryText']
KCCAns = df['KCCAns']
for i,j in zip(query_text,KCCAns):
    print(i,'\n',j,'\n\n')
```

```python
punctuation_char = string.punctuation + ''
# punctuation_char = punctuation_char.replace('-','')
querry_clean=[]
for i in query_text:
    i=i.translate(str.maketrans('','',punctuation_char))
    w = i.lower().split()
    temp = []
    for ws in w:
        if ws not in useless_words:
            temp.append(ws)
    querry_clean.append(temp)
```
